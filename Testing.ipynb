Can testing help? Assuming testing is followed by isolation or quarantine, for high risk cases.

```python
def testing(graph_1, N_C_1, start):
    Rand = []
    nodes = len(list(graph_1.neighbors(0)))
    for i in range(1,nodes + 1):
        if N_C_1[i] == 'red':
            Rand.append(i)
            L_1 = list(graph_1.neighbors(i))[1:]
            for l in L_1:
                N_C_1[l] = 'green'
    for j in range(start):
        R = random.randint(1, nodes) # nodes
        while R in Rand:
            #print('R: ' + str(R))
            R = random.randint(1, nodes) # nodes
        Rand.append(R)
        N_C_1[R] = 'green'
        L_2 = list(graph_1.neighbors(R))[1:]
        for l in L_2:
            N_C_1[l] = 'green'
        
    #print(N_C_1)
```

We start random testing, and isolating or quarantining of high risk people prevents the spread.

```python
testing(graph_1, N_C_1, start = 2)
plt.figure(figsize=(8, 8))
nx.draw(graph_1, node_size=60, alpha=0.5, node_color=N_C_1, with_labels=False)#, pos
plt.draw()
plt.show()
```
