```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
def infection(rand, graph_1, start, low, high):
    for i in range(start): # start
        R = random.randint(low, high) # 1, nodes
        while R in rand:
            #print('R: ' + str(R))
            R = random.randint(low, high) # 1, nodes
        rand.append(R)
        L_1 = list(graph_1.neighbors(R))[1:]
        #print(L_1)
        if L_1 != []:
            infection(rand, graph_1, start = start, low = min(L_1), high = max(L_1))
```

```python
def network_plot(nodes, level, start):
    graph_1 = nx.balanced_tree(nodes, level) # nodes, level
    roles_1 = [0] * graph_1.number_of_nodes()
    roles_1[0] = 1
    N_N_1 = graph_1.number_of_nodes()
    print(N_N_1)

    rand=[]

    infection(rand, graph_1, start = start, low = 1, high = nodes) # nodes  

    #print(rand)
    
    for r in rand:
        roles_1[r] = 1
    
    #print(roles_1)
    
    N_C_1 = []

    for r_1 in range(N_N_1):
        if roles_1[r_1] == 1:
            N_C_1.append('red')
        else:
            if roles_1[r_1] == 0:
                N_C_1.append('blue')

    #print(N_C_1)

    return graph_1, N_C_1
```

Here we see the spread of COVID - 19 pre lockdown.

```python
graph_1, N_C_1 = network_plot(nodes = 8, level = 2, start = 2)
plt.figure(figsize=(8, 8))
nx.draw(graph_1, node_size=60, alpha=0.5, node_color=N_C_1, with_labels=False)#, pos
plt.draw()
plt.show()
```

Here we see the spread of COVID - 19 post lockdown.
